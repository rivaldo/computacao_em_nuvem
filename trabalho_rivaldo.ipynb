### 1.1 Quantos filmes estão disponíveis no dataset?

```python
print (f"Quantitdade de filmes no dataset: {len(filmes['titulo_ano_lancamento'])} filmes.")
```

```python
nome_filmes = []
filme_anos = []

for a in range(len(filmes['id_filme'])):
    
    nome_filme = filmes['titulo_ano_lancamento'][a].split(',')
    nome_filme = nome_filme[0].split('(')
    print(nome_filme[1])
    nome_filmes.append(nome_filme[1]) 
    
    ano_filme = filmes['titulo_ano_lancamento'][a].split(',')
    ano_filme = ano_filme[1].split(')')
    filme_anos.append(ano_filme[0])

    
len(filme_anos)
```

```python
dados = {
    'Nome do Filme': nome_filmes,
    'Ano de Producao': filme_anos
}

lista_de_filmes = pd.DataFrame(dados)
lista_de_filmes.head()
```

### 1.2 Qual o nome dos 5 filmes com melhor media de avaliação?

```python
media_5 = df_avaliacao.groupby(df_avaliacao['titulo_ano_lancamento'])
```

```python
print(f'Os 5 filmes melhores avaliados sao: {media_5.avaliacao.mean().sort_values(ascending=False).head(5).round(2)}')
```
